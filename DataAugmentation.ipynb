```python
from PIL import Image
import numpy as np
import cv2
# Esta linha é específica para o Google Colab
from google.colab.patches import cv2_imshow
```

```python
# Caminho para a imagem local
caminho_imagem = '/content/batrachia.jpg'  # Atualize com o caminho real para a sua imagem
```

```python
# Carrega a imagem usando Pillow
img = Image.open(caminho_imagem)
img = np.array(img)

# Se a imagem estiver em modo RGBA (com canal alpha), converte para RGB para evitar erros ao exibir com o OpenCV
if img.shape[-1] == 4:
    img = cv2.cvtColor(img, cv2.COLOR_RGBA2RGB)
else:
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

# Exibe a imagem com a função adaptada para o Colab
cv2_imshow(img)
```

```python
from keras.preprocessing.image import ImageDataGenerator # Importa biblioteca de data augmentation
# Cria o objeto de data agumentation
aug = ImageDataGenerator(height_shift_range=0.2, width_shift_range=0.3, rotation_range=45, zoom_range=[0.8, 1.0])
```

```python
# Cria uma nova dimensão
imagem = np.array(img)
imagem = np.expand_dims(imagem, axis=0)

print(imagem.shape)
```

```python
# Realiza data augmentation com a imagem passada
imageGen = aug.flow(imagem, batch_size = 1)
```

```python
# Imprime imagens geradas
for i in range(5):
  next = imageGen.next()
  imgPrint = np.array(next)
  cv2_imshow(imgPrint[0,:,:,:])
```
